### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
import matplotlib.pyplot as plt
```

```python
#Metodo para calcular el todal de ventas(suma el todal de la culumna precio)
def calc_total_ventas(listaPrecios):
  total_ventas = sum(listaPrecios)
  return total_ventas


#Calculamos total de ventas de cada tienda y creamos una tupla donde almacenamos los valores
total_ventas_tienda = calc_total_ventas(tienda['Precio'].tolist())
total_ventas_tienda2 = calc_total_ventas(tienda2['Precio'].tolist())
total_ventas_tienda3 = calc_total_ventas(tienda3['Precio'].tolist())
total_ventas_tienda4 = calc_total_ventas(tienda4['Precio'].tolist())

tiendas_total_ventas = (total_ventas_tienda, total_ventas_tienda2, total_ventas_tienda3, total_ventas_tienda4)
```

# 2. Ventas por categoría

```python



#Con este metodo camos a calcular cuantos productos se han vendido por categoria
def calc_ventas_por_categoria(listaCategorias):
  counter = {}
  for categoria in listaCategorias:
    if categoria in counter :
      counter[categoria] += 1
    else:
      counter[categoria] = 1

  return counter


##Calculamos total ventas por categoria para cada tienda
ventas_por_categoria_tienda = calc_ventas_por_categoria(tienda['Categoría del Producto'].tolist())
ventas_por_categoria_tienda2 = calc_ventas_por_categoria(tienda2['Categoría del Producto'].tolist())
ventas_por_categoria_tienda3 = calc_ventas_por_categoria(tienda3['Categoría del Producto'].tolist())
ventas_por_categoria_tienda4 = calc_ventas_por_categoria(tienda4['Categoría del Producto'].tolist())

#Tupla de ventas
tiendas_ventas_por_categoria = (ventas_por_categoria_tienda, ventas_por_categoria_tienda2, ventas_por_categoria_tienda3, ventas_por_categoria_tienda4)
```

# 3. Calificación promedio de la tienda


```python
#Metodo para calcular calificaciones
def calcPromedio(listaCalificaciones):
  promedio = sum(listaCalificaciones) / len(listaCalificaciones)
  return promedio

#calculamos el promedio para cada tienda
promedio_tienda = calcPromedio(tienda['Calificación'].tolist())
promedio_tienda2 = calcPromedio(tienda2['Calificación'].tolist())
promedio_tienda3 = calcPromedio(tienda3['Calificación'].tolist())
promedio_tienda4 = calcPromedio(tienda4['Calificación'].tolist())

#Craemos la tupla
tiendas_promedios_calificacion= (promedio_tienda, promedio_tienda2, promedio_tienda3, promedio_tienda4)
```

# 4. Productos más y menos vendidos

```python
#Calcular productos mas vendidos
def calc_ventas_por_producto(listaProductos):
  counter = {}
  for producto in listaProductos:
    if producto in counter :
      counter[producto] += 1
    else:
      counter[producto] = 1

  return counter


#Productos mas y menos vendidos por cada tienda
ventas_por_producto_tienda = calc_ventas_por_producto(tienda['Producto'].tolist())
ventas_por_producto_tienda2 = calc_ventas_por_producto(tienda2['Producto'].tolist())
ventas_por_producto_tienda3 = calc_ventas_por_producto(tienda3['Producto'].tolist())
ventas_por_producto_tienda4 = calc_ventas_por_producto(tienda4['Producto'].tolist())

#Tupla de ventas por producto de cada tienda
tiendas_ventas_por_producto = (ventas_por_producto_tienda, ventas_por_producto_tienda2, ventas_por_producto_tienda3, ventas_por_producto_tienda4)
```

# 5. Envío promedio por tienda

```python
##Metodo para calcular el envio promedio

def calc_envio_promedio(listaEnvios):
  envio_promedio = sum(listaEnvios) / len(listaEnvios)
  return envio_promedio

#calculamos el promedio de envio para cada tienda y creamos una tupla
envio_promedio_tienda = calc_envio_promedio(tienda['Costo de envío'].tolist())
envio_promedio_tienda2 = calc_envio_promedio(tienda2['Costo de envío'].tolist())
envio_promedio_tienda3 = calc_envio_promedio(tienda3['Costo de envío'].tolist())
envio_promedio_tienda4 = calc_envio_promedio(tienda4['Costo de envío'].tolist())

tiendas_envio_promedio = (envio_promedio_tienda, envio_promedio_tienda2, envio_promedio_tienda3, envio_promedio_tienda4)
```

# Graficos


```python
tiendas = ("Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4")

# Crear gráfico
plt.figure(figsize=(10, 6))
plt.bar(tiendas, tiendas_total_ventas, color='skyblue')
plt.title('Total de Ventas por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Total de Ventas')

plt.ticklabel_format(style='plain', axis='y')

plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python

fig, axs = plt.subplots(1, 4, figsize=(20, 5))

for i in range(4):
    categorias = list(tiendas_ventas_por_categoria[i].keys())
    ventas = list(tiendas_ventas_por_categoria[i].values())

    # Grafico circular
    axs[i].pie(ventas, labels=categorias, autopct='%1.1f%%', startangle=90)
    axs[i].set_title(f'Ventas por Categoría\n({tiendas[i]})')

    # Calcular maximo y minimo
    categoria_max = max(tiendas_ventas_por_categoria[i], key=tiendas_ventas_por_categoria[i].get)
    categoria_min = min(tiendas_ventas_por_categoria[i], key=tiendas_ventas_por_categoria[i].get)

    print(f"{tiendas[i]} - Más vendido: {categoria_max} ({tiendas_ventas_por_categoria[i][categoria_max]} ventas)")
    print(f"{tiendas[i]} - Menos vendido: {categoria_min} ({tiendas_ventas_por_categoria[i][categoria_min]} ventas)")
    print('************************************************')


plt.tight_layout()
plt.show()


```

```python

# Barra horizontal
plt.barh(tiendas, tiendas_promedios_calificacion, color='skyblue')

# Títulos y etiquetas
plt.title('Calificación Promedio por Tienda')
plt.xlabel('Calificación Promedio')
plt.ylabel('Tienda')

# Mostrar el gráfico
plt.show()
```

```python
fig, axs = plt.subplots(4, 1, figsize=(20,100))  # 4 filas, 1 columna

for i in range(4):
    categorias = list(tiendas_ventas_por_producto[i].keys())
    ventas = list(tiendas_ventas_por_producto[i].values())

    # Gráfico de barras horizontal
    axs[i].barh(categorias, ventas, color='skyblue')
    axs[i].set_title(f'Ventas por Producto\n({tiendas[i]})')
    axs[i].set_xlabel('Ventas')
    axs[i].set_ylabel('Producto')

    axs[i].invert_yaxis()

    # Calcular máximo y mínimo
    categoria_max = max(tiendas_ventas_por_producto[i], key=tiendas_ventas_por_producto[i].get)
    categoria_min = min(tiendas_ventas_por_producto[i], key=tiendas_ventas_por_producto[i].get)

    print(f"{tiendas[i]} - Más vendido: {categoria_max} ({tiendas_ventas_por_producto[i][categoria_max]} ventas)")
    print(f"{tiendas[i]} - Menos vendido: {categoria_min} ({tiendas_ventas_por_producto[i][categoria_min]} ventas)")
    print('************************************************')

plt.tight_layout()
plt.show()
```

```python
#Envio promedio por tienda
plt.bar(tiendas, tiendas_envio_promedio, color='skyblue')

plt.title('Total Costo Envio Promedio por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Costo de Envio Promedio')

# Mostrar el gráfico
plt.show()
```
